```python
import pandas as pd
import json
```

```python
df = pd.read_csv('elephantfootsizeandage.csv')
# df = pd.read_csv('roman.csv')
df.head()
```

```python
lines = []
for fname,gr in df.groupby('filename'):
    line = [fname]
    js = gr.sort_values('region_id')['region_shape_attributes'].tolist()
    for j in js:
        d = json.loads(j)
        if 'cx' in d:
            line.extend([int(d["cx"]),int(d["cy"])])
    lines.append(line)
```

```python
pd.DataFrame(lines).to_csv('elelandfootage.csv', index=False, header=None)
# pd.DataFrame(lines).to_csv('roman2.csv', index=False, header=None)
```
